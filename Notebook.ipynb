```python
import numpy as np
import pandas as pd
from scipy.sparse import csr_matrix
```

```python
books = pd.read_csv('BX-Books.csv', sep=";", error_bad_lines=False, encoding='latin-1')
```

```python
books.columns
```

```python
books = books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher']]
books.rename(columns={'Book-Title':'title', 'Book-Author': 'author', 'Year-Of-Publication': 'year', 'Publisher': 'publisher'}, inplace=True)
books.head(1)
```

```python
users = pd.read_csv('BX-Users.csv', sep=";", error_bad_lines=False, encoding='latin-1')
users.head(2)
```

```python
users.rename(columns={'User-ID':'user_id', 'Location': 'location', 'Age': 'age'}, inplace=True)
users.head(1)
```

```python
ratings = pd.read_csv('BX-Book-Ratings.csv', sep=";", error_bad_lines=False, encoding='latin-1')
ratings.head(2)
```

```python
ratings.rename(columns={'User-ID':'user_id', 'Book-Rating': 'rating'}, inplace=True)
ratings.head(1)
```

```python
users.shape
```

```python
ratings.shape
```

```python
ratings['user_id'].value_counts()
```

```python
#number of unique users
ratings['user_id'].value_counts().shape
```

```python
#users who rated atleast 200 books
ratings['user_id'].value_counts()>200
```

```python
x = ratings['user_id'].value_counts()>200

#extracting rows only with True values
x[x]
```

```python
x[x].shape
```

```python
y = x[x].index
y
```

```python
#keeping only those records where users are common in y

ratings = ratings[ratings['user_id'].isin(y)]
ratings.head()
```

```python
ratings.shape
```

```python
# Joining ratings and books on the basis on ISBN

ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books
```

```python
ratings_with_books.groupby('title')['rating'].count()
```

```python
ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
number_rating.head(2)
```

```python
number_rating.rename(columns={'rating': 'number of rating'}, inplace=True)
number_rating.head(1)
```

```python
final_rating = ratings_with_books.merge(number_rating, on='title')
final_rating
```

```python
final_rating = final_rating[final_rating['number of rating']>=50]
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id', 'title'], inplace=True)
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
book_pivot
```

```python
book_sparse = csr_matrix(book_pivot)
type(book_sparse)
```

```python
# KNN is used for classification but NearestNeighbors is used for clustering

from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1, -1), n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index==book_name)[0][0]
    distances, suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)
    
    for i in range(len(suggestions)):
        if i==0:
            print("The suggestions for", book_name, "are :")
        if not i:
            print(book_pivot.index[suggestions[i]])
```

```python
recommend_book('Animal Farm')
```
